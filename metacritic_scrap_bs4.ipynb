```python
id_list = []
url_list = []
img_list = []
title_list = []
platform_list = []
date_list = []
summary_list = []
metascore_list = []
userscore_list = []
genre_list = []
```

```python
START_PAGE = 2
GENRE = 'Wrestling'
```

```python
pages = range(START_PAGE)
idx = START_PAGE

for page in pages:
    page = requests.get('https://www.metacritic.com/browse/games/genre/date/wrestling/all?view=detailed&page='+str(idx-1), headers={'User-Agent': 'Mozilla/5.0'})
    soup = BeautifulSoup(page.text, 'html.parser')
    sleep(randint(2, 4))
    idx -= 1
    
    print("Current page: {}".format(idx))
    
    tables = soup.find_all('table', {"class": "clamp-list"})
    table_idx = 0
    for table in tables:
        rows = table.find_all('tr')
        row_idx = 0
        
        print("Current table: {}".format(table_idx))
        table_idx += 1

        for row in rows:
            class_name = row.get('class')
            if(class_name == ['spacer']):
                continue
            
            id_game = row.find('input', {"class": "clamp-summary-expand"}).get('id')
            url = 'https://www.metacritic.com' + row.find('a', {"class": "title"}).get('href')
            img = row.find('td', {"class": "clamp-image-wrap"}).find('img').get('src')
            title = row.find('a', {"class": "title"}).get_text()
            platform = row.find('span', {"class": "data"}).get_text().strip()
            release_date = row.find('div', {"class": "clamp-details"}).find_all('span')[2].get_text()
            summary = row.find('div', {"class": "summary"}).get_text().strip()
            metascore = row.find('div', {"class": "clamp-metascore"}).find('div', {"class" : 'metascore_w'}).get_text()
            userscore = row.find('div', {"class": "clamp-userscore"}).find('div', {"class" : 'metascore_w'}).get_text()

            id_list.append(id_game)
            url_list.append(url)
            img_list.append(img)
            title_list.append(title)
            platform_list.append(platform)
            date_list.append(release_date)
            summary_list.append(summary)
            metascore_list.append(metascore)
            userscore_list.append(userscore)
            genre_list.append(GENRE)
            
            print("Current row: {}".format(row_idx))
            row_idx += 1
```

```python
df = pd.DataFrame({
    'id': id_list,
    'url': url_list,
    'image': img_list, 
    'title': title_list,
    'platform': platform_list,
    'release_date': date_list,
    'summary': summary_list,
    'metascore': metascore_list,
    'userscore': userscore_list,
    'genre': genre_list,
})
```
