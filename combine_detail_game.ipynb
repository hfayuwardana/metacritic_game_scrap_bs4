```python
import pandas as pd
import os
import numpy as np
import re
from datetime import datetime
```

```python
list_df = []
```

# Read CSV

```python
directory = 'GITHUB/detail_fix/'

for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    df = pd.read_csv(f) # membaca file CSV
    list_df.append(df)
```

```python
directory2 = 'GITHUB/detail_fix2/'

for filename in os.listdir(directory2):
    f = os.path.join(directory2, filename)
    df = pd.read_csv(f) # membaca file CSV
    list_df.append(df)
```

# Menggabungkan beberapa dataframe menjadi satu

```python
# merge beberapa dataframe menjadi satu dataframe
df = pd.concat(list_df)
```

```python
df = df.iloc[:, 1:]
```

# Bersihkan dataset

## a. Kolom 'is_online'

```python
def clean_player(x):
    if x is np.nan:
        return 0
    else:
        return 1
```

```python
df['is_online'] = df['player'].apply(clean_player)
```

```python
df['is_online'] = df['is_online'].astype(object)
```

```python
df['is_online'].unique()
```
