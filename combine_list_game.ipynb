```python
import pandas as pd
import os
import numpy as np
import re
from datetime import datetime
```

# Membaca file CSV

```python
list_df = []
directory = 'GITHUB/fix/'

for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    df = pd.read_csv(f) # membaca file CSV
    list_df.append(df)
```

# Menggabungkan beberapa dataframe menjadi satu

```python
# merge beberapa dataframe menjadi satu dataframe
df = pd.concat(list_df)
```

```python
df.index = np.arange(0, len(df))
```

```python
df = df.iloc[:, 1:]
```

# Bersihkan nilai untuk setiap kolom pada dataframe

## a. Kolom 'release_date'

```python
def get_date(x):
    pattern = r'(\w+) (\d+), (\d+)'
    res = re.match(pattern, x)
    month, day, year = res.groups()
    
    correct_date = datetime.strptime(month[:3] + " " + day + " " + year, '%b %d %Y')
    return correct_date
```

```python
df['release_date'] = df['release_date'].apply(get_date)
```

```python
df.head()
```

## b. Kolom 'platform'

```python
df['platform'].unique() # before
```

```python
def remove_spaces(x):
    x = " ".join(x.split())
    return x
```

```python
df['platform'] = df['platform'].apply(remove_spaces)
```

```python
df['platform'].unique() # after
```

## c. Kolom 'userscore'

```python
df['userscore'].unique()
```

```python
def clean_userscore(x):
    if x == 'tbd':
        x = '0'
    return x
```

```python
df['userscore'] = df['userscore'].apply(clean_userscore)
```

```python
df['userscore'].unique()
```

```python
df['userscore'] = df['userscore'].astype('float')
```

## d. Kolom 'id'

```python
df['id'] = df['id'].astype('object')
```
